#### Problema 1
 #### Explicación y DespejePlanteamiento: Identificamos la incógnita. "Un número" lo representaremos con la letra $x$. La frase "más el número nueve" indica una suma ($+9$) y "es igual a 21" establece la igualdad ($= 21$).Ecuación:$$x + 9 = 21$$Despeje: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Hallar el número desconocido
# resultado_total = 21
# numero_conocido = 9
resultado_total=int(input("Ingrese el resultado total: "))
numero_conocido=int(input("Ingrese el número conocido: "))
# Despeje: x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número buscado es: {x}")
```

#### Problema 2
 #### Las edades de Pedro y AnaAquí manejamos una relación entre dos variables que podemos reducir a una sola.Paso 1: Explicación y DespejePlanteamiento: * Edad de Ana: $x$Edad de Pedro: $2x$ (porque es el doble que la de Ana).La suma de ambos: $36$.Ecuación:$$x + 2x = 36$$Despeje:Sumamos los términos semejantes: $3x = 36$El 3 que multiplica a la $x$ pasa dividiendo:$$x = \frac{36}{3}$$Solución:$x = 12$ (Edad de Ana)$2x = 2(12) = 24$ (Edad de Pedro)

```python
# Problema 2: Edades de Pedro y Ana
# suma_edades = 36
suma_edades=int(input("Ingrese la suma de las edades: "))
# La ecuación es x + 2x = 36 -> 3x = 36
# Dividimos el total entre 3 para hallar la unidad (edad de Ana)
edad_ana = suma_edades / 3
edad_pedro = edad_ana * 2

print(f"Ana tiene {int(edad_ana)} años")
print(f"Pedro tiene {int(edad_pedro)} años")
```
